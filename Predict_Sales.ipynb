# Scenarios - DATA ANALYSIS PART


### Total Revenue by Category

```python
cursor.execute("""select category, (unitprice * quantity) as total_revenue from orders
Group by category
Order by total_revenue
""");
revenue_by_category=cursor.fetchall()
revenue_by_category_df= pd.DataFrame(revenue_by_category, columns=['Category', 'TotalSales'])
revenue_by_category_df
```

```python
grouped_df = revenue_by_category_df.copy()
grouped_df['Category'] = grouped_df['Category'].str.title()
grouped_df = grouped_df.groupby('Category').sum().reset_index()

# Plot with nicer visuals
plt.figure()
bars = plt.bar(grouped_df['Category'], grouped_df['TotalSales'])

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.title('Revenue by Category', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

### Monthly Sales Trend

```python
import calendar
```

```python
cursor.execute("""PRAGMA table_info(orders);""")
cursor.fetchall()
```

```python
cursor.execute("""SELECT
    strftime('%m', REPLACE(orderdate, '/', '-')) AS month,
    SUM(quantity * unitprice) AS sales
FROM orders
GROUP BY month;""")
total_sales_by_month=pd.DataFrame(cursor.fetchall(),columns=["Month","Sales"])
month_map = {f"{i:02d}": calendar.month_name[i] for i in range(1, 13)}
total_sales_by_month["Month"]=total_sales_by_month["Month"].map(month_map)
total_sales_by_month=total_sales_by_month.dropna(subset=["Month"])
total_sales_by_month
```

```python
plt.figure()
bars=plt.bar(total_sales_by_month["Month"],total_sales_by_month["Sales"])
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.xlabel("Months", fontsize=14)
plt.ylabel("Sales", fontsize=14)
plt.title("Monthly Sales",fontsize=18)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

# Data Modelling part - Data Science

```python
cursor.execute("select * from orders;")
orders_df=pd.DataFrame(cursor.fetchall(),columns=["OrderId","CustomerName","Product","Category","Quantity","UnitPrice","OrderDate","PaymentMethod"])
orders_df=orders_df.drop(columns=["OrderId"])
orders_df["Category"]=orders_df["Category"].str.lower()
# orders_df['OrderDate'] = pd.to_datetime(orders_df['OrderDate'], format='%Y/%m/%d', errors='coerce')
orders_df
```

```python
def clean_dates(df, date_col):
    df_clean = df.copy()
    
    # Step 1: Replace / with -
    df_clean[date_col] = df_clean[date_col].str.replace('/', '-')
    
    # Step 2: Parse each row properly
    def parse_date(s):
        s = str(s)
        # if starts with year (2023-...)
        if s[:4].isdigit() and int(s[:4]) > 1900:
            return pd.to_datetime(s, format='%Y-%m-%d', errors='coerce')
        # else assume day-month-year
        return pd.to_datetime(s, format='%d-%m-%Y', errors='coerce')
    
    df_clean[date_col] = df_clean[date_col].apply(parse_date)
    df_clean['Year'] = df_clean[date_col].dt.year
    df_clean['Month'] = df_clean[date_col].dt.month_name()
    df_clean['Day'] = df_clean[date_col].dt.day
    df_clean['Weekday'] = df_clean[date_col].dt.day_name()
    
    return df_clean
```

```python
orders_df["Amount"]=orders_df["Quantity"] * orders_df["UnitPrice"]
cleaned_date=clean_dates(orders_df,"OrderDate")
orders_df = pd.concat([orders_df, cleaned_date[['Year','Month', 'Day', 'Weekday']]], axis=1)
orders_df
```

```python
orders_df['Weekend'] = orders_df['Weekday'].apply(lambda d: 1 if d in ['Friday','Saturday', 'Sunday'] else 0)
orders_df.columns
```

```python
encoding_columns=["Category","PaymentMethod"]
encoded_df=pd.get_dummies(orders_df,columns=encoding_columns,drop_first=True)
encoded_df
```

```python
features=encoded_df.drop(columns=["CustomerName","Product","OrderDate"])
features['Month'] = pd.to_datetime(features['Month'], format='%B').dt.month
weekday_map = {'Monday':0, 'Tuesday':1, 'Wednesday':2, 'Thursday':3, 'Friday':4, 'Saturday':5, 'Sunday':6}
features['Weekday']=features["Weekday"].map(weekday_map)
features
```

```python
data=features
data["WeekendHighValue"]= ((features["Weekend"]==1) & (features["Amount"]>=1000).astype(int))
data
```

```python
features=data.drop(columns=["WeekendHighValue"])
output=data["WeekendHighValue"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,output,test_size=0.2,random_state=12)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
print(f"Accuracy Score: {accuracy_score(y_test,prediction)}")
print("Classification Report:")
print(classification_report(y_test, prediction))
print("Confusion Matrix:")
print(confusion_matrix(y_test, prediction))
```
